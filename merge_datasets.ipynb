# Strip 'kkst' from project_id

```python
old_train['project_id'] = old_train['project_id'].map(lambda x: x.lstrip('kkst'))
```

```python
old_test['project_id'] = old_test['project_id'].map(lambda x: x.lstrip('kkst'))
```

# Merge/Join with the new data

```python
# using simple inner join - Pandas merge() defaults to an “inner” merge operation
# only the joined rows - probably be smaller than the real dataset - old_train
merged_train = pd.merge(old_train,
                 new_dataset[['id', 'photo', 'pledged', 'state', 'staff_pick', 'is_starrable', 'static_usd_rate', 'usd_pledged', 'creator', 'category', 'profile', 'spotlight', ]],
                 left_on='project_id',
                 right_on='id')
merged_train.head()
```

```python
# using simple left join - shape will be same as left table i.e. 'old_train' - with joined and not joined rows.
merged_train_two = pd.merge(old_train,
                            new_dataset[['id', 'photo', 'pledged', 'state', 'staff_pick', 'is_starrable', 'static_usd_rate', 'usd_pledged', 'creator', 'category', 'profile', 'spotlight', ]],
                            left_on='project_id',
                            right_on='id', 
                            how='left')
merged_train_two.head()
```
